## IMPORTATION DES BIBLIOTHEQUES

```python
# Importation des bibliothèques nécessaires
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
```

## CHARGEMENT DES DONNES IRIS

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

## DIVISION DES DONNEES EN ENSEMBLE D'ENTRAINNEMENT ET DE TEST

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 50)
```

## NORMALISATION DES DONNEES

```python
# Création d'un objet StandardScaler
norm = StandardScaler()

# Normalisation des données d'entrainnement
X_train = norm.fit_transform(X_train)

# Normalisation des données de test
X_test = norm.fit_transform(X_test)
```

## CREATION ET ENTRAINNEMENT DU MODEL LASSO

```python
# Création du modèle Lasso avec un paramètre de régularisation
lasso = Lasso(alpha = 0.1)

# Entrainnement du modèle
lasso.fit(X_train, y_train)
```

## PREDICTION SUR L'ENSEMBLE DE TEST

```python
# Prédiction des cibles pour les données de test
y_pred = lasso.predict(X_test)
y_pred;
```

## EVALUATION DU MODELE

```python
# Calcul et affichage du MSE
mse = mean_squared_error(y_test, y_pred)
print(f'Erreur quadratique moyenne (MSE) = {mse}')
```
